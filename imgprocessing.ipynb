```python
 for v in g:   
    path = "dataset/" + str(v) + "/"
    c=0
    for image_path in os.listdir(path):
        #angle = random.uniform(-15, 15)
        count=0
        q=[]
        # create the full input path and read the file
        input_path = os.path.join(path, image_path)
        c+=1
        if c>5500:
            image = cv2.imread(input_path)
            #image = sk.transform.rotate(image, angle)
            #image = imutils.rotate(image, angle)
            #image = cv2.resize(image, (28,28), interpolation = cv2.INTER_AREA)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            blur = cv2.GaussianBlur(gray, (blurValue, blurValue), 0)
            thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, blockSize=15, C=5)
            #thresh = imutils.rotate(thresh, angle)
            ret, thresh = cv2.threshold(thresh, threshold, 255, cv2.THRESH_BINARY)
            im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
            for cnt in contours:
                q.append(cv2.contourArea(cnt))
            q.sort(reverse=True)
            for cnt in contours:
                if cv2.contourArea(cnt)<q[0]/10:
                    f,y,w,h = cv2.boundingRect(cnt)
                    for i in range(y,y+h):
                        for j in range(f,f+w):
                            thresh[i][j]=0
            kernel = np.ones((2,2), np.uint8)
            thresh = cv2.dilate(thresh, kernel, iterations=1)
            #ret, thresh = cv2.threshold(blur, threshold, 255, cv2.THRESH_BINARY_INV)
            for i in range(45):
                for j in range(45):
                    t[count].append(int(thresh[i][j]/255))
                    count+=1
            y_test.append(v)
            if c>6000:
                break
        else:
            image = cv2.imread(input_path)
            #image = sk.transform.rotate(image, angle)
            #image = cv2.resize(image, (28,28), interpolation = cv2.INTER_AREA)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            blur = cv2.GaussianBlur(gray, (blurValue, blurValue), 0)
            #ret, thresh = cv2.threshold(blur, threshold, 255, cv2.THRESH_BINARY_INV)
            thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, blockSize=15, C=5)
            #thresh = imutils.rotate(thresh, angle)
            ret, thresh = cv2.threshold(thresh, threshold, 255, cv2.THRESH_BINARY)
            im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
            for cnt in contours:
                q.append(cv2.contourArea(cnt))
            q.sort(reverse=True)
            for cnt in contours:
                if cv2.contourArea(cnt)<q[0]/12:
                    f,y,w,h = cv2.boundingRect(cnt)
                    for i in range(y,y+h):
                        for j in range(f,f+w):
                            thresh[i][j]=0
            kernel = np.ones((2,2), np.uint8)
            thresh = cv2.dilate(thresh, kernel, iterations=1)
            for i in range(45):
                for j in range(45):
                    l[count].append(int(thresh[i][j]/255))
                    count+=1
            y_train.append(v)
```
